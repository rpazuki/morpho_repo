# Brusselator model

> The equations are as follows

<h3><center>$\partial_t u = D_u (\partial_x^2 + \partial_y^2)u + A - (B+1)u + u^2v$</center></h3>

<h3><center>$\partial_t v = D_v (\partial_x^2 + \partial_y^2)v + Bu - u^2 v$</center></h3>

With parameters  :
- $D_u = 0.001, \quad D_v=1/900=0.00111111$
- $A = 20$
- $B = 400$
- $n = (100, 100)$
- $L = (0.005, 0.005)$
- $dt = 10^{-4}$

```python
n = (64, 64)
L = (.05, .05)

D_u = .001
D_v = 0.1#.002
print(D_u/D_v)
A = 1#9.002250562640661

Ds = np.array([D_u, D_v])
print(Ds)
#Bs = [1.13049110640674-.2, 1.13049110640674+.001, 1.13049110640674+.2]
Bs = [10, 15, 20, 25, 35]#np.linspace(1, 40, num=10, endpoint=True)
sols = []
for B in Bs:
    #B = B + 1.13049110640674
    print(f"B={B}")
    #c0 = np.random.normal(scale=.1, size=(2, n[0], n[1]))
    c0 = np.zeros((2, n[0], n[1]))
    c0[0, :,:] += A
    c0[1, :,:] += B/A
    c0[0, :,:] += np.eye(n[0], n[1]) * .001
    c0[1, :,:] += np.eye(n[0], n[1]) * .001


    f_args = (A, B)
    t = np.linspace(0, 50, 100)

    dt = 0.00001

    sol = integrate(c0, t, dt, n, L, Ds, Brusselator, f_args)
    index =0
    _ = plt.figure(figsize=(8,4))
    plt.subplot(121)
    plt.imshow(sol[0,:,:, index], cmap='BuGn', interpolation='gaussian')
    plt.colorbar(shrink=.75)
    plt.subplot(122)
    plt.imshow(sol[1,:,:, index], cmap='BuGn', interpolation='gaussian')
    plt.colorbar(shrink=.75)
    plt.show()

    index =-1
    _ = plt.figure(figsize=(8,4))
    plt.subplot(121)
    vmin = np.min(sol[0, :, :, 0])
    vmax = np.max(sol[0, :, :, 0])
    plt.imshow(sol[0,:,:, index], 
               cmap='BuGn',
               vmin=vmin,
               vmax=vmax,
               interpolation='gaussian')
    plt.colorbar(shrink=.75)
    plt.subplot(122)
    vmin = np.min(sol[1, :, :, 0])
    vmax = np.max(sol[1, :, :, 0])
    plt.imshow(sol[1,:,:, index],
               cmap='BuGn',
               vmin=vmin,
               vmax=vmax,
               interpolation='gaussian')
    plt.colorbar(shrink=.75)
    plt.show()
    
    sols += [sol]
```
