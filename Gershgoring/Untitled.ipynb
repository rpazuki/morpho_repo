```python
import os
os.environ["XLA_PYTHON_CLIENT_MEM_FRACTION"]="0.1"

import jax.numpy as jnp
from jax import grad, jit, vmap
from jax import random

import matplotlib.pyplot as plt
```

```python
@jit
def act(x, K, n):
    return 1/(1 + jnp.power(K/x, n))
@jit
def inh(x, K, n):
    return 1/(1 + jnp.power(x/K,n))

```

```python

xs = jnp.arange(0, 8.0, 0.01)
derivative_act = grad(act, argnums=0)
derivative_inh = grad(inh, argnums=0)
```

```python
np.power((4+1)/(4-1), 1/4)
```

```python
_ = plt.figure(figsize=(12,4))
plt.subplot(121)
plt.plot(xs, [derivative_act(x, .5, 4) for x in xs], label=f"k=0.5")
plt.plot(xs, [derivative_act(x, 1, 4) for x in xs], label=f"k=1")
plt.plot(xs, [derivative_act(x, 2, 4) for x in xs], label=f"k=2")
plt.plot(xs, [derivative_act(x, 3, 4) for x in xs], label=f"k=4")
plt.legend();plt.grid()
plt.subplot(122)
plt.plot(xs, [derivative_inh(x, .5, 4) for x in xs], label=f"k=0.5")
plt.plot(xs, [derivative_inh(x, 1, 4) for x in xs], label=f"k=1")
plt.plot(xs, [derivative_inh(x, 2, 4) for x in xs], label=f"k=2")
plt.plot(xs, [derivative_inh(x, 3, 4) for x in xs], label=f"k=4")
plt.legend();plt.grid();plt.show()
_ = plt.figure(figsize=(12,4))
plt.subplot(121)
plt.plot(xs, [derivative_act(x, .5, 2) for x in xs], label=f"k=0.5")
plt.plot(xs, [derivative_act(x, 1, 2) for x in xs], label=f"k=1")
plt.plot(xs, [derivative_act(x, 2, 2) for x in xs], label=f"k=2")
plt.plot(xs, [derivative_act(x, 3, 2) for x in xs], label=f"k=4")
plt.legend();plt.grid()
plt.subplot(122)
plt.plot(xs, [derivative_inh(x, .5, 2) for x in xs], label=f"k=0.5")
plt.plot(xs, [derivative_inh(x, 1, 2) for x in xs], label=f"k=1")
plt.plot(xs, [derivative_inh(x, 2, 2) for x in xs], label=f"k=2")
plt.plot(xs, [derivative_inh(x, 3, 2) for x in xs], label=f"k=4")
plt.legend();plt.grid()
```
