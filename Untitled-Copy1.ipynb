```python
from numba import cuda, float32
import numpy as np
import math

# @cuda.jit
# def RHS_GPU(c0, dt, n, L, Ds, f_args, dc):
#     # Define an array in the shared memory
#     # The size and type of the arrays must be known at compile time
#     #sdc = cuda.shared.array(shape=(TPB, TPB), dtype=float32)    

#     #x, y = cuda.grid(2)
    
#     #tx = cuda.threadIdx.x
#     #ty = cuda.threadIdx.y
#     #bpg = cuda.gridDim.x    # blocks per grid
#     z = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x
#     x = cuda.blockIdx.y * cuda.blockDim.y + cuda.threadIdx.y
#     y = cuda.blockIdx.z * cuda.blockDim.z + cuda.threadIdx.z
#     if z >= c0.shape[0] and x >= c0.shape[1] and y >= c0.shape[2]:
#         # Quit if (x, y) is outside of valid dc boundary
#         return
       
#     #dc = 0
    
#     # Periodic boundary condition
#     i_prev = (x - 1) % n[0]
#     i_next = (x + 1) % n[0]

#     j_prev = (y - 1) % n[1]
#     j_next = (y + 1) % n[1]
    
#     A, B = f_args
    
    
#     u = c0[0, x, y]
#     v = c0[1, x, y]
#     u2 = u**2
#     u2v = u2 * v
#     if z == 0:
#         fuv = A - (B + 1) * u + u2v
#     else:
#         fuv = B * u - u2v    
            
#     dc[z, x, y] = dt*((Ds[z] * n[z] / L[z]) * ( c0[z, i_prev, y]
#         + c0[z, i_next, y]
#         + c0[z, x, j_prev]
#         + c0[z, x, j_next]
#         - 4.0 * c0[z, x, y]
#     ) + fuv)  


@cuda.jit
def Brusselator_GPU(c0, f_args, z):
    A, B = f_args
    u = c0[0]
    v = c0[1]
    u2 = u**2
    u2v = u2 * v
    if z == 0:
        return A - (B + 1) * u + u2v
    else:
        return B * u - u2v

def create_GPU(f):    
    @cuda.jit
    def RHS_GPU(c0, dt, n, L, Ds, f_args, dc):

        z = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x
        x = cuda.blockIdx.y * cuda.blockDim.y + cuda.threadIdx.y
        y = cuda.blockIdx.z * cuda.blockDim.z + cuda.threadIdx.z
        if z >= c0.shape[0] or x >= c0.shape[1] or y >= c0.shape[2]:
            # Quit if (x, y) is outside of valid dc boundary
            return

        # Periodic boundary condition
        i_prev = (x - 1) % n[0]
        i_next = (x + 1) % n[0]

        j_prev = (y - 1) % n[1]
        j_next = (y + 1) % n[1]


        fuv = f(c0[:, x, y], f_args, z)        

        dc[z, x, y] = dt*((Ds[z] * n[z] / L[z]) * ( c0[z, i_prev, y]
            + c0[z, i_next, y]
            + c0[z, x, j_prev]
            + c0[z, x, j_next]
            - 4.0 * c0[z, x, y]
        ) + fuv) 
    return RHS_GPU

@cuda.jit
def add_GPU(c, dc):
    z = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x
    x = cuda.blockIdx.y * cuda.blockDim.y + cuda.threadIdx.y
    y = cuda.blockIdx.z * cuda.blockDim.z + cuda.threadIdx.z
    if z >= c.shape[0] or x >= c.shape[1] or y >= c.shape[2]:
        # Quit if (x, y) is outside of valid dc boundary
        return
    c[z, x, y] += dc[z, x, y]
    
    
@cuda.jit
def assign_GPU(cs, c, index):
    z = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x
    x = cuda.blockIdx.y * cuda.blockDim.y + cuda.threadIdx.y
    y = cuda.blockIdx.z * cuda.blockDim.z + cuda.threadIdx.z
    if z >= c.shape[0] or x >= c.shape[1] or y >= c.shape[2]:
        # Quit if (x, y) is outside of valid dc boundary
        return
    cs[z, x, y, index] = c[z, x, y]
```

```python
n = (128, 128)
L = (0.05, 0.05)


D_u = .001
D_v = .020
Ds = np.array([D_u, D_v])
A = 2
B = 3

c0 = np.random.normal(scale=.1, size=(2, n[0], n[1]))
c0[0, :, :] += A
c0[1, :, :] += B/A
#c0 = data[:, :, :, 0]

f_args = (A, B)
t = np.linspace(0, 128, 64)
#t = np.linspace(0, 1, 2)

dt = 0.001

dc = np.zeros((c0.shape[0], c0.shape[1], c0.shape[2]))
ret = np.zeros((c0.shape[0], c0.shape[1], c0.shape[2], t.shape[0]))

threadsperblock = (2, 16, 16)
blockspergrid_x = math.ceil(c0.shape[0] / threadsperblock[0])
blockspergrid_y = math.ceil(c0.shape[1] / threadsperblock[1])
blockspergrid_z = math.ceil(c0.shape[2] / threadsperblock[2])

blockspergrid = (blockspergrid_x, blockspergrid_y, blockspergrid_z)

d_dc = cuda.to_device(dc)
d_n = cuda.to_device(n)
d_L = cuda.to_device(L)
d_Ds = cuda.to_device(Ds)
d_f_args = cuda.to_device(f_args)
d_ret = cuda.to_device(ret)
```

```python
RHS_GPU = create_GPU(Brusselator_GPU)
```

```python
d_c = cuda.to_device(c0)
t_c = t[0]
assign_GPU[blockspergrid, threadsperblock](d_ret, d_c, 0)
for t_i, t_next in enumerate(t[1:]):
    while t_c < t_next:        
        RHS_GPU[blockspergrid, threadsperblock](
                d_c,
                dt,
                d_n, 
                d_L, 
                d_Ds, 
                d_f_args, 
                d_dc)
        add_GPU[blockspergrid, threadsperblock](d_c, d_dc)
        t_c += dt
    assign_GPU[blockspergrid, threadsperblock](d_ret, d_c, t_i+1)
        
```
